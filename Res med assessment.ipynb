```
from pyspark.sql import SparkSession, functions as F
from urllib.request import urlopen

spark = SparkSession.builder.getOrCreate()

url = 'https://s3-us-west-2.amazonaws.com/ph-coding-challenges-ci8dghf3p0/clinical.json'
jsonData = urlopen(url).read().decode('utf-8')
rdd = spark.sparkContext.parallelize([jsonData])
df = spark.read.json(rdd)
```

```
url = 'https://s3-us-west-2.amazonaws.com/ph-coding-challenges-ci8dghf3p0/environmental.json'
jsonData = urlopen(url).read().decode('utf-8')
rdd = spark.sparkContext.parallelize([jsonData])
df1 = spark.read.json(rdd)
```

```
df1 = df1.withColumnRenamed("station", "env_station") \
          .withColumnRenamed("hour", "env_hour") 
```
